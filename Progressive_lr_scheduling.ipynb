# Experiment 1

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, random_split
import time
import matplotlib.pyplot as plt
import numpy as np
from tqdm.auto import tqdm
import copy
from sklearn.metrics import accuracy_score, f1_score
from transformers import ViTImageProcessor, ViTForImageClassification
from datasets import load_dataset
import wandb  # Added for experiment tracking

# Set device and seed for reproducibility
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.manual_seed(42)

class PlantSpeciesDataset(Dataset):
    """Custom dataset for jbarat/plant_species to work with PyTorch DataLoader"""
    def __init__(self, hf_dataset, processor, split='train'):
        self.dataset = hf_dataset[split]
        self.processor = processor

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        item = self.dataset[idx]
        image = item['image']
        label = item['label']

        # Process image for ViT
        inputs = self.processor(images=image, return_tensors="pt")
        pixel_values = inputs['pixel_values'].squeeze()

        return pixel_values, label

def get_progressive_layer_lr_groups(model, base_lr, layer_decay_factor):
    """
    Create parameter groups with progressive layer-specific learning rates.
    Earlier layers get smaller learning rates, later layers get larger rates.
    """
    param_groups = []

    # Embedding layer (captures very generic features)
    embedding_params = list(model.vit.embeddings.parameters())
    embedding_lr = base_lr * (layer_decay_factor ** 0)  # Lowest learning rate
    param_groups.append({'params': embedding_params, 'lr': embedding_lr, 'layer_name': 'embedding'})

    # Transformer blocks (progressively increasing learning rates)
    num_layers = len(model.vit.encoder.layer)
    for i, block in enumerate(model.vit.encoder.layer):
        # Progressively increase learning rate for later layers
        layer_lr = base_lr * (layer_decay_factor ** (i + 1))
        param_groups.append({
            'params': block.parameters(),
            'lr': layer_lr,
            'layer_name': f'transformer_block_{i}'
        })

    # Head layer (classification layer - needs most adaptation)
    head_params = list(model.classifier.parameters())
    head_lr = base_lr * (layer_decay_factor ** (num_layers + 1))  # Highest learning rate
    param_groups.append({'params': head_params, 'lr': head_lr, 'layer_name': 'head'})

    return param_groups

class LRMonitorCallback:
    """Callback to monitor and log learning rates during training"""
    def __init__(self, optimizer):
        self.optimizer = optimizer

    def on_batch_end(self):
        """Log learning rates after each batch"""
        # Log each layer's learning rate if progressive LR is used
        for i, param_group in enumerate(self.optimizer.param_groups):
            if 'layer_name' in param_group:
                wandb.log({f"lr/{param_group['layer_name']}": param_group['lr']}, commit=False)
            else:
                wandb.log({f"lr/group_{i}": param_group['lr']}, commit=False)

        # Log average LR
        lrs = [param_group['lr'] for param_group in self.optimizer.param_groups]
        wandb.log({"lr/avg": np.mean(lrs)}, commit=False)

class SchedulerExperiment:
    """Main experiment class for testing different learning rate schedulers"""

    def __init__(self, config):
        """
        Initialize the experiment with configuration

        Args:
            config: Dictionary with experiment configuration
        """
        self.config = config
        self.batch_size = config.get('batch_size', 32)
        self.num_epochs = config.get('num_epochs', 10)
        self.base_lr = config.get('base_lr', 0.0001)
        self.layer_decay_factor = config.get('layer_decay_factor', 1.2)
        self.weight_decay = config.get('weight_decay', 0.05)
        self.scheduler_type = config.get('scheduler_type', 'cyclic')
        self.scheduler_params = config.get('scheduler_params', {})
        self.use_wandb = config.get('use_wandb', True)

        # Initialize Weights & Biases if enabled
        if self.use_wandb:
            wandb.init(
                project="vit-progressive-lr-schedulers",
                name=config.get('experiment_name', f"{self.scheduler_type}_lr_{self.base_lr}"),
                config=config
            )

        # Load model processor
        self.processor = ViTImageProcessor.from_pretrained('google/vit-base-patch16-224-in21k')

        # Load and prepare data
        self.prepare_data()

        # Initialize models
        self.init_models()

        # Results storage
        self.results = {
            'progressive': {
                'train_loss': [], 'val_loss': [],
                'train_acc': [], 'val_acc': [],
                'time_per_epoch': []
            },
            'uniform': {
                'train_loss': [], 'val_loss': [],
                'train_acc': [], 'val_acc': [],
                'time_per_epoch': []
            }
        }

    def prepare_data(self):
        """Load and prepare jbarat/plant_species dataset"""
        print("Preparing jbarat/plant_species dataset...")

        try:
            # Load the dataset from Hugging Face
            dataset = load_dataset("jbarat/plant_species")

            # Create custom datasets
            train_dataset = PlantSpeciesDataset(dataset, self.processor, split='train')

            # Get number of classes
            self.num_classes = len(set(dataset['train']['label']))

            # Split train dataset into train and validation
            train_size = int(0.8 * len(train_dataset))
            val_size = len(train_dataset) - train_size
            self.train_dataset, self.val_dataset = random_split(train_dataset, [train_size, val_size])

            # Test dataset
            self.test_dataset = self.val_dataset

        except Exception as e:
            print(f"Error loading jbarat/plant_species dataset: {e}")
            print("\nPlease ensure the dataset is available or manually download and organize it.")
            raise

        # Create data loaders
        self.train_loader = DataLoader(self.train_dataset, batch_size=self.batch_size, shuffle=True, num_workers=2)
        self.val_loader = DataLoader(self.val_dataset, batch_size=self.batch_size, shuffle=False, num_workers=2)
        self.test_loader = DataLoader(self.test_dataset, batch_size=self.batch_size, shuffle=False, num_workers=2)

    def init_models(self):
        """Initialize pre-trained ViT models for the experiment"""
        print("Initializing models...")

        # Load pre-trained ViT model from Hugging Face
        base_model = ViTForImageClassification.from_pretrained(
            'google/vit-base-patch16-224-in21k',
            num_labels=self.num_classes,
            ignore_mismatched_sizes=True  # Important when changing head size
        )

        # Create two identical models for our comparison
        self.model_progressive = copy.deepcopy(base_model).to(device)
        self.model_uniform = copy.deepcopy(base_model).to(device)

    def create_scheduler(self, optimizer, scheduler_type, is_progressive=False):
        """
        Create learning rate scheduler based on specified type
        """
        total_steps = len(self.train_loader) * self.num_epochs
        scheduler_name = f"{'progressive' if is_progressive else 'uniform'}_{scheduler_type}"

        if scheduler_type == 'cyclic':
            # CyclicLR scheduler
            step_size_up = self.scheduler_params.get('step_size_up', total_steps // (2 * self.num_epochs))
            max_lr = self.scheduler_params.get('max_lr', self.base_lr * 10)
            mode = self.scheduler_params.get('mode', 'triangular')

            if is_progressive:
                # For progressive LR, scale max_lr for each layer based on layer_decay_factor
                base_max_lrs = []
                for group in optimizer.param_groups:
                    # Calculate max_lr proportionally to the base_lr of each layer
                    layer_max_lr = group['lr'] * (max_lr / self.base_lr)
                    base_max_lrs.append(layer_max_lr)

                return optim.lr_scheduler.CyclicLR(
                    optimizer,
                    base_lr=[group['lr'] for group in optimizer.param_groups],
                    max_lr=base_max_lrs,
                    step_size_up=step_size_up,
                    mode=mode,
                    cycle_momentum=False
                )
            else:
                return optim.lr_scheduler.CyclicLR(
                    optimizer,
                    base_lr=self.base_lr,
                    max_lr=max_lr,
                    step_size_up=step_size_up,
                    mode=mode,
                    cycle_momentum=False
                )

        elif scheduler_type == 'cosine':
            # CosineAnnealingLR scheduler
            T_max = self.scheduler_params.get('T_max', total_steps)
            eta_min = self.scheduler_params.get('eta_min', 0)

            return optim.lr_scheduler.CosineAnnealingLR(
                optimizer,
                T_max=T_max,
                eta_min=eta_min
            )

        elif scheduler_type == 'cosine_restart':
            # CosineAnnealingWarmRestarts scheduler
            T_0 = self.scheduler_params.get('T_0', total_steps // self.num_epochs)
            T_mult = self.scheduler_params.get('T_mult', 1)
            eta_min = self.scheduler_params.get('eta_min', 0)

            return optim.lr_scheduler.CosineAnnealingWarmRestarts(
                optimizer,
                T_0=T_0,
                T_mult=T_mult,
                eta_min=eta_min
            )
        else:
            raise ValueError(f"Unsupported scheduler type: {scheduler_type}")

    def train_epoch(self, model, optimizer, scheduler=None, mode='progressive'):
        """Train for one epoch"""
        model.train()
        total_loss = 0.0
        correct = 0
        total = 0
        start_time = time.time()

        # Create callback for LR logging if using wandb
        lr_callback = LRMonitorCallback(optimizer) if self.use_wandb else None

        for batch_idx, (inputs, targets) in enumerate(tqdm(self.train_loader, desc=f"Training ({mode})")):
            inputs, targets = inputs.to(device), targets.to(device)

            optimizer.zero_grad()

            # Forward pass with HuggingFace model returns loss if labels are provided
            outputs = model(pixel_values=inputs, labels=targets)
            loss = outputs.loss

            loss.backward()
            optimizer.step()

            # Step the scheduler - important for cyclical schedulers after each batch
            if scheduler:
                if self.scheduler_type in ['cyclic', 'cosine_restart']:
                    scheduler.step()
                elif self.scheduler_type == 'cosine' and batch_idx == len(self.train_loader) - 1:
                    # For cosine annealing, step once per epoch
                    scheduler.step()

            # Log learning rates if using wandb
            if lr_callback:
                lr_callback.on_batch_end()

            # Calculate accuracy
            pred = outputs.logits.argmax(dim=1)
            total += targets.size(0)
            correct += (pred == targets).sum().item()

            total_loss += loss.item() * inputs.size(0)

            # Log batch metrics if using wandb
            if self.use_wandb:
                wandb.log({
                    f"{mode}/batch_loss": loss.item(),
                    f"{mode}/batch_acc": (pred == targets).float().mean().item(),
                    "batch": batch_idx + len(self.train_loader) * (self.current_epoch)
                })

        epoch_time = time.time() - start_time
        epoch_loss = total_loss / total
        epoch_acc = correct / total

        return epoch_loss, epoch_acc, epoch_time

    def validate(self, model, mode='progressive'):
        """Validate the model"""
        model.eval()
        total_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            for inputs, targets in tqdm(self.val_loader, desc=f"Validating ({mode})"):
                inputs, targets = inputs.to(device), targets.to(device)

                # Forward pass with HuggingFace model
                outputs = model(pixel_values=inputs, labels=targets)
                loss = outputs.loss

                # Calculate accuracy
                pred = outputs.logits.argmax(dim=1)
                total += targets.size(0)
                correct += (pred == targets).sum().item()

                total_loss += loss.item() * inputs.size(0)

        val_loss = total_loss / total
        val_acc = correct / total

        return val_loss, val_acc

    def run_experiment(self):
        """Run the full experiment comparing progressive and uniform learning rates"""
        print("\n" + "=" * 50)
        print(f"STARTING EXPERIMENT: {self.scheduler_type.upper()} Scheduler, Base LR: {self.base_lr}")
        print("=" * 50)

        # Setup optimizers
        # 1. Progressive learning rate
        param_groups_progressive = get_progressive_layer_lr_groups(
            self.model_progressive, self.base_lr, self.layer_decay_factor
        )
        optimizer_progressive = optim.AdamW(param_groups_progressive, weight_decay=self.weight_decay)
        self.optimizer_progressive = optimizer_progressive  # Store for visualization

        # 2. Uniform learning rate
        optimizer_uniform = optim.AdamW(
            self.model_uniform.parameters(), lr=self.base_lr, weight_decay=self.weight_decay
        )
        self.optimizer_uniform = optimizer_uniform  # Store for visualization

        # Setup schedulers
        scheduler_progressive = self.create_scheduler(optimizer_progressive, self.scheduler_type, is_progressive=True)
        scheduler_uniform = self.create_scheduler(optimizer_uniform, self.scheduler_type, is_progressive=False)

        # Train and evaluate each model
        best_val_acc_progressive = 0.0
        best_val_acc_uniform = 0.0

        for epoch in range(self.num_epochs):
            self.current_epoch = epoch  # Store current epoch for batch numbering
            print(f"\nEpoch {epoch+1}/{self.num_epochs}")
            print("-" * 30)

            # 1. Progressive learning rate model
            train_loss, train_acc, epoch_time = self.train_epoch(
                self.model_progressive, optimizer_progressive, scheduler_progressive, mode='progressive'
            )
            val_loss, val_acc = self.validate(self.model_progressive, mode='progressive')

            # Save results
            self.results['progressive']['train_loss'].append(train_loss)
            self.results['progressive']['train_acc'].append(train_acc)
            self.results['progressive']['val_loss'].append(val_loss)
            self.results['progressive']['val_acc'].append(val_acc)
            self.results['progressive']['time_per_epoch'].append(epoch_time)

            print(f"Progressive LR - Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, "
                  f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}, Time: {epoch_time:.2f}s")

            if val_acc > best_val_acc_progressive:
                best_val_acc_progressive = val_acc

            # Log to wandb if enabled
            if self.use_wandb:
                wandb.log({
                    "progressive/train_loss": train_loss,
                    "progressive/train_acc": train_acc,
                    "progressive/val_loss": val_loss,
                    "progressive/val_acc": val_acc,
                    "progressive/epoch_time": epoch_time,
                    "epoch": epoch + 1
                })

            # 2. Uniform learning rate model
            train_loss, train_acc, epoch_time = self.train_epoch(
                self.model_uniform, optimizer_uniform, scheduler_uniform, mode='uniform'
            )
            val_loss, val_acc = self.validate(self.model_uniform, mode='uniform')

            # Save results
            self.results['uniform']['train_loss'].append(train_loss)
            self.results['uniform']['train_acc'].append(train_acc)
            self.results['uniform']['val_loss'].append(val_loss)
            self.results['uniform']['val_acc'].append(val_acc)
            self.results['uniform']['time_per_epoch'].append(epoch_time)

            print(f"Uniform LR - Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, "
                  f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.4f}, Time: {epoch_time:.2f}s")

            if val_acc > best_val_acc_uniform:
                best_val_acc_uniform = val_acc

            # Log to wandb if enabled
            if self.use_wandb:
                wandb.log({
                    "uniform/train_loss": train_loss,
                    "uniform/train_acc": train_acc,
                    "uniform/val_loss": val_loss,
                    "uniform/val_acc": val_acc,
                    "uniform/epoch_time": epoch_time,
                    "epoch": epoch + 1
                })

        # Calculate performance metrics and time savings
        self._evaluate_and_report_final_results(best_val_acc_progressive, best_val_acc_uniform)

        # Finish wandb run if enabled
        if self.use_wandb:
            wandb.finish()

        return self.results

    def _evaluate_and_report_final_results(self, best_val_acc_progressive, best_val_acc_uniform):
        """Evaluate models on test set and report final results"""
        # Calculate training time differences
        total_time_progressive = sum(self.results['progressive']['time_per_epoch'])
        total_time_uniform = sum(self.results['uniform']['time_per_epoch'])
        time_reduction_percent = (1 - total_time_progressive / total_time_uniform) * 100

        print("\n" + "=" * 50)
        print("EXPERIMENT RESULTS:")
        print(f"Best validation accuracy (Progressive LR): {best_val_acc_progressive:.4f}")
        print(f"Best validation accuracy (Uniform LR): {best_val_acc_uniform:.4f}")
        print(f"Accuracy difference: {(best_val_acc_progressive - best_val_acc_uniform) * 100:.2f}%")
        print(f"Total training time (Progressive LR): {total_time_progressive:.2f}s")
        print(f"Total training time (Uniform LR): {total_time_uniform:.2f}s")
        print(f"Training time reduction: {time_reduction_percent:.2f}%")
        print("=" * 50)

        # Test the models on the test set
        self.model_progressive.eval()
        self.model_uniform.eval()

        progressive_preds = []
        uniform_preds = []
        true_labels = []

        with torch.no_grad():
            for inputs, targets in tqdm(self.test_loader, desc="Testing models"):
                inputs, targets = inputs.to(device), targets.to(device)

                # Get predictions
                outputs_progressive = self.model_progressive(pixel_values=inputs)
                outputs_uniform = self.model_uniform(pixel_values=inputs)

                pred_progressive = outputs_progressive.logits.argmax(dim=1)
                pred_uniform = outputs_uniform.logits.argmax(dim=1)

                # Collect predictions and true labels
                progressive_preds.extend(pred_progressive.cpu().numpy())
                uniform_preds.extend(pred_uniform.cpu().numpy())
                true_labels.extend(targets.cpu().numpy())

        # Calculate final metrics
        progressive_acc = accuracy_score(true_labels, progressive_preds)
        uniform_acc = accuracy_score(true_labels, uniform_preds)
        progressive_f1 = f1_score(true_labels, progressive_preds, average='macro')
        uniform_f1 = f1_score(true_labels, uniform_preds, average='macro')

        print("\nFinal Test Metrics:")
        print(f"Progressive LR - Accuracy: {progressive_acc:.4f}, F1 Score: {progressive_f1:.4f}")
        print(f"Uniform LR - Accuracy: {uniform_acc:.4f}, F1 Score: {uniform_f1:.4f}")

        # Log final metrics to wandb if enabled
        if self.use_wandb:
            wandb.log({
                "final/progressive_test_acc": progressive_acc,
                "final/uniform_test_acc": uniform_acc,
                "final/progressive_f1": progressive_f1,
                "final/uniform_f1": uniform_f1,
                "final/accuracy_difference": progressive_acc - uniform_acc,
                "final/training_time_reduction": time_reduction_percent
            })

    def visualize_results(self, save_path=None):
        """Visualize the experiment results with plots"""
        if not self.results['progressive']['train_loss']:
            print("No results to visualize. Run the experiment first.")
            return

        # Create plots for training/validation metrics
        fig, axs = plt.subplots(2, 2, figsize=(15, 10))
        epochs = range(1, self.num_epochs + 1)

        # Plot loss curves
        axs[0, 0].plot(epochs, self.results['progressive']['train_loss'], 'b-', label='Progressive - Train')
        axs[0, 0].plot(epochs, self.results['progressive']['val_loss'], 'b--', label='Progressive - Val')
        axs[0, 0].plot(epochs, self.results['uniform']['train_loss'], 'r-', label='Uniform - Train')
        axs[0, 0].plot(epochs, self.results['uniform']['val_loss'], 'r--', label='Uniform - Val')
        axs[0, 0].set_title('Loss')
        axs[0, 0].set_xlabel('Epoch')
        axs[0, 0].set_ylabel('Loss')
        axs[0, 0].legend()
        axs[0, 0].grid(True)

        # Plot accuracy curves
        axs[0, 1].plot(epochs, [acc * 100 for acc in self.results['progressive']['train_acc']], 'b-',
                       label='Progressive - Train')
        axs[0, 1].plot(epochs, [acc * 100 for acc in self.results['progressive']['val_acc']], 'b--',
                      label='Progressive - Val')
        axs[0, 1].plot(epochs, [acc * 100 for acc in self.results['uniform']['train_acc']], 'r-',
                      label='Uniform - Train')
        axs[0, 1].plot(epochs, [acc * 100 for acc in self.results['uniform']['val_acc']], 'r--',
                      label='Uniform - Val')
        axs[0, 1].set_title('Accuracy')
        axs[0, 1].set_xlabel('Epoch')
        axs[0, 1].set_ylabel('Accuracy (%)')
        axs[0, 1].legend()
        axs[0, 1].grid(True)

        # Plot training time
        axs[1, 0].plot(epochs, self.results['progressive']['time_per_epoch'], 'b-', label='Progressive')
        axs[1, 0].plot(epochs, self.results['uniform']['time_per_epoch'], 'r-', label='Uniform')
        axs[1, 0].set_title('Time per Epoch')
        axs[1, 0].set_xlabel('Epoch')
        axs[1, 0].set_ylabel('Time (seconds)')
        axs[1, 0].legend()
        axs[1, 0].grid(True)

        # Plot the layer-specific learning rates
        layer_names = []
        layer_lrs = []

        # Extract layer names and initial LRs
        for param_group in self.optimizer_progressive.param_groups:
            if 'layer_name' in param_group:
                layer_names.append(param_group['layer_name'])
                layer_lrs.append(param_group['lr'])

        # Plot learning rates by layer
        axs[1, 1].bar(range(len(layer_names)), layer_lrs)
        axs[1, 1].set_title('Initial Learning Rates per Layer')
        axs[1, 1].set_ylabel('Learning Rate')
        axs[1, 1].set_xticks(range(len(layer_names)))
        axs[1, 1].set_xticklabels(layer_names, rotation=45, ha='right')
        axs[1, 1].grid(True, axis='y')

        plt.tight_layout()

        # Save figure if path is provided
        if save_path:
            plt.savefig(save_path)
        else:
            plt.savefig(f'{self.scheduler_type}_lr_results.png')

        plt.show()

# Define experiment configurations
def run_experiments():
    """Run multiple experiments with different schedulers"""
    experiment_configs = [
        # Cyclic LR experiment
        {
            'experiment_name': 'cyclic_progressive_vs_uniform',
            'batch_size': 16,
            'num_epochs': 5,
            'base_lr': 0.00001,
            'layer_decay_factor': 1.1,
            'weight_decay': 0.05,
            'scheduler_type': 'cyclic',
            'scheduler_params': {
                'max_lr': 0.0001,
                'step_size_up': 2000,
                'mode': 'triangular'
            }
        },
        # Cosine Annealing experiment
        {
            'experiment_name': 'cosine_progressive_vs_uniform',
            'batch_size': 16,
            'num_epochs': 5,
            'base_lr': 0.00001,
            'layer_decay_factor': 1.1,
            'weight_decay': 0.05,
            'scheduler_type': 'cosine',
            'scheduler_params': {
                'eta_min': 1e-6
            }
        },
        # Cosine Annealing with Warm Restarts experiment
        {
            'experiment_name': 'cosine_restart_progressive_vs_uniform',
            'batch_size': 16,
            'num_epochs': 5,
            'base_lr': 0.00001,
            'layer_decay_factor': 1.1,
            'weight_decay': 0.05,
            'scheduler_type': 'cosine_restart',
            'scheduler_params': {
                'T_0': 1000,
                'T_mult': 2,
                'eta_min': 1e-6
            }
        }
    ]

    # Run each experiment
    for config in experiment_configs:
        print(f"\nRunning experiment: {config['experiment_name']}")
        experiment = SchedulerExperiment(config)
        results = experiment.run_experiment()
        experiment.visualize_results(f"{config['experiment_name']}_results.png")

if __name__ == "__main__":
    # Run all experiments

    run_experiments()

    # Or run a single experiment
    # config = {
    #     'experiment_name': 'cyclic_progressive_vs_uniform',
    #     'batch_size': 16,
    #     'num_epochs': 2,
    #     'base_lr': 0.00001,
    #     'layer_decay_factor': 1.1,
    #     'weight_decay': 0.05,
    #     'scheduler_type': 'cyclic',
    #     'scheduler_params': {
    #         'max_lr': 0.0001,
    #         'step_size_up': 2000,
    #         'mode': 'triangular'
    #     }
    # }
    # experiment = SchedulerExperiment(config)
    # results = experiment.run_experiment()
    # experiment.visualize_results()
```
